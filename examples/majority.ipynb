```python
from synet.process import MajorityProcess
from synet.networks import random_network

from matplotlib import pyplot as plt
```

```python
majority = MajorityProcess()
```

```python
net = random_network(n_events=1000)
```

```python
for _ in range(30):
    net = random_network(n_events=10000)
    plt.plot(majority.simulate_dt(net, dt=2000, n_sim=128, n_jobs=16))
    plt.xscale('log')
plt.show()
```
