```python
from matplotlib import pyplot as plt
import numpy as np

from synet.networks import random_network
from synet.process import MajorityProcess
from synet.utils import apply_process, apply_measures
from synet.visualization import plot_process_results, plot_measure_results
from synet.visualization import plot_process_v_measure, plot_pvm_dt, plot_bootstrap_sim
```

```python
n_networks = 100
process_dt = 250
n_agents = 100
n_events = 5000
np.random.seed(1287364)
```

```python
majority = MajorityProcess()
```

```python
%%time
networks = [random_network(n_agents=n_agents, n_events=n_events, n_community=3) for _ in range(n_networks)]
```

```python
%%time
process_results = apply_process(networks, process=majority, dt=process_dt, n_sim=500, n_jobs=1)
```

```python
%%time
measure_results = apply_measures(networks, max_dt=100, n_jobs=32)
```

```python
#plot_bootstrap_sim(process_results, measure_results)
```

```python
plot_process_results(process_results)
```

```python
plot_process_v_measure(process_results, measure_results)
```

```python
plot_pvm_dt(process_results, measure_results)
```
